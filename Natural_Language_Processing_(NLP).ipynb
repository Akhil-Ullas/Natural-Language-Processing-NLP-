# **Stemming**

```python
words=['running','goes','history','english','studies','teach','walking','congratulations']
```

**Porter Stemmer**

```python
# porterstemmer/word stemmer

from nltk.stem import PorterStemmer
ps=PorterStemmer()

for i in words:
  print(i+' ---> '+ps.stem(i))
```

**RegxStemmer**

```python
# RegexStemmer

from nltk.stem import RegexpStemmer
rs=RegexpStemmer("ing$|s$|sh$")

for i in words:
  print(i+' ---> '+rs.stem(i))
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
le=WordNetLemmatizer()

for i in words:
  print(i+' ---> '+le.lemmatize(i))
```

# 1. **One hot encoding :**

eg:

**One hot encoding**

**Example 1**

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import string #text cleaning
```

```python
document=['The cat sat on the mat',
          'The dog sat on the log',
          'Cats and dogs are pets']
```

```python
words=[word.lower().strip(string.punctuation)
for doc in document for word in doc.split()]
vocabulary=sorted(set(words)) #alphabetical order
print(vocabulary)

'''
Loops through each document and each word
convert word to lowercase, remove punctuation, produce a list of cleaned tokens
'''
```

```python
encoder=OneHotEncoder(sparse_output=False)
one_hot_vector=encoder.fit_transform(np.array(vocabulary).reshape(-1,1))

word_to_onehot={vocabulary[i]:one_hot_vector[i]
                for i in range(len(vocabulary))}

for word,vector in word_to_onehot.items():
  print(f'word:\t{word},\tOne_Hot_Encoding:\t{vector}')
```

# **3. TF - IDF**

TF -> Term Frequency

* number of rep of words in sentence / number of words in sentence

IDF -> Inverse Document Frequency

* log(e) (number of sentence / number of sentence containing the word)

**Example:**

s1: White pasta

s2: White sauce

s3: White sauce pasta

**TF**:
                     
                    voc       s1    s2    s3

                    White     1/2   1/2   1/3

                    Pasta     1/2   0     1/3

                    Sauce     0     1/2   1/3


**IDF**:

                    words                   IDF

                    White                   log (3/3)

                    Pasta                   log (3/2)

                    Sauce                   log (3/2)


                           

Equation for TF-IDF:

                        TF-IDF = TF x IDF

```python
text=text.split()
```

```python
print(text[:20])
```

```python
def freqplot(words,top_n=5):
  word_freq=FreqDist(words)
  labels=[element[0] for element in word_freq.most_common(top_n)]
  counts=[element[1] for element in word_freq.most_common(top_n)]
  plt.figure(figsize=(15,5))
  plt.title('Most Frequent Words in the text including stop words')
  plt.ylabel('Count')
  plt.xlabel('Word')
  plot=plt.bar(labels,counts)
  return plot
freqplot(text,20)
plt.show()
```

```python
word_freq=FreqDist(text)
word_freq.most_common(20)
```

```python
ranks=[]
freqs=[]
for rank, (word,freq) in enumerate(word_freq.most_common()):
  ranks.append(rank+1)
  freqs.append(freq)
```

```python
plt.figure(figsize=(20,7))
plt.loglog(freqs,ranks)
plt.xlabel('Frequency')
plt.ylabel('Rank')
plt.title('Frequency-Rank Plot')
plt.show()
```
